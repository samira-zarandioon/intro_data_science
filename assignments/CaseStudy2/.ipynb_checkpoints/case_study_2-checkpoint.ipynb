## (i)

```python
import pandas
temp_countries = pandas.read_csv('./TEMP.csv')
temp_countries_gt_1900 = temp_countries[temp_countries['Date'] > '1900-01-01']
temp_countries_gt_1900_agg = temp_countries_gt_1900.groupby(['Country'])['Monthly AverageTemp'].agg({'min_temp': min, 'max_temp': max})
temp_countries_gt_1900_agg['diff_temp'] = abs(temp_countries_gt_1900_agg['min_temp'] - temp_countries_gt_1900_agg['max_temp'])
temp_countries_gt_1900_agg = temp_countries_gt_1900_agg.drop(['min_temp', 'max_temp'], axis=1)
temp_countries_gt_1900_agg_20_largest = temp_countries_gt_1900_agg.nlargest(20, 'diff_temp')
```

```python
temp_countries_gt_1900_agg_20_largest
```

```python
p = temp_countries_gt_1900_agg_20_largest.plot(y='diff_temp',kind='bar')
```

## (ii)

```python
UStemp = temp_countries[(temp_countries['Date'] >= '1900-01-01') & (temp_countries['Country']=='United States')].copy()
```

### (ii) a)

```python
UStemp['TempInFahrenheit'] = 1.8 * UStemp['Monthly AverageTemp'] + 32
```

### (ii) b)

```python
UStemp['Date'] = pandas.to_datetime(UStemp['Date'])
UStemp['Year'] = UStemp['Date'].dt.year
```

```python
import numpy as np
yearly_UStemp = UStemp.groupby(['Year'])['Monthly AverageTemp'].agg({"Yearly AverageTemp" : np.mean}).reset_index()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
p = yearly_UStemp.plot(x='Year', y='Yearly AverageTemp', label="Yearly Average US Temp")
```

### (ii) c)

```python
yearly_UStemp_diff = abs(yearly_UStemp['Yearly AverageTemp'] - yearly_UStemp['Yearly AverageTemp'].shift())
maxindex = np.argmax(yearly_UStemp_diff)
y1 = yearly_UStemp.loc[maxindex-1]['Year'].astype(int)
y2 = yearly_UStemp.loc[maxindex]['Year'].astype(int)
max_diff = yearly_UStemp_diff[maxindex]
print("Max temp difference is between year %s and %s and is equal to %s." % (y1, y2, max_diff))
```

## (iii)

```python
import pandas
temp_cities = pandas.read_csv('./CityTemp.csv')
temp_cities_gt_1900 = temp_cities[temp_cities['Date'] > '1900-01-01']
temp_cities_gt_1900_agg = temp_cities_gt_1900.groupby(['Country', 'City'])['Monthly AverageTemp'].agg({"min_temp": min, "max_temp": max})
temp_cities_gt_1900_agg['diff_temp'] = abs(temp_cities_gt_1900_agg['min_temp'] - temp_cities_gt_1900_agg['max_temp'])
temp_cities_gt_1900_agg = temp_cities_gt_1900_agg.drop(['max_temp', 'min_temp'], axis = 1)
temp_cities_gt_1900_agg_20_largest = temp_cities_gt_1900_agg.nlargest(20, 'diff_temp')
```

```python
p = temp_cities_gt_1900_agg_20_largest.plot(y='diff_temp',kind='bar')
```

## (iv) 

The observation of the two graphs shows that they have approximately the same slope as we can see that both histograms have a maximum value around 48o and the minimum value around 35o. Another remark is that the countries and cities that have the max average temperature are located in the same continent like China (in the country, City graph) and Kazakhstan (in the Country graph) are in Asia, and the lowest values are for Europe as Turkey and Arminia.
